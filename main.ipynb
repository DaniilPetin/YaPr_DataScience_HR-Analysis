# Проект: HR-аналитика

В распоряжении имеются данные о сотрудниках компании "Работа с заботой". Необходимо помочь отделу HR-аналитики компании избежать финансовых потерь и оттока сотрудников при помощи инструментов машинного обучения и Data Science. 
Компания предоставила данные с характеристиками сотрудников компании, в том числе - уровень удовлетворённости сотрудника работой в компании. 

Перед исполнителем проекта ставится **2 основные задачи**:
1. Построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика. 
2. Построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.

В ходе работы над **первой задачей** планируется выполнить следующие этапы:
   1. Загрузка данных и изучение общей информации;
   2. Предобработка данных: проверка данные на наличие пропусков и дубликатов. Проверка корректности типов данных у каждого признака. Устранение недочетов при необходимости;
   3. Исследовательский анализ данных;
   4. Подготовка данных для обучения;
   5. Обучение моделей и выбор лучшей при помощи метрики SMAPE;
   6. Оформление выводов.

В ходе работы над **второй задачей** планируется выполнить следующие этапы:
   1. Загрузка данных и изучение общей информации;
   2. Предобработка данных;
   3. Исследовательский анализ данных;
   4. Добавления нового входного признака - удовлетворенности работой, полученной по результатом работы первой модели;
   5. Подготовка данных для обучения;
   6. Обучение моделей, подбор гиперпараметров и выбор лучшей при помощи метрики ROC-AUC;
   7. Оформление выводов

По результатам выполнения обеих задач будет сформулирован **общий вывод**.

Данные для работы находятся в нескольких файлах:
- 1 задача:
    - Тренировочная выборка: train_job_satisfaction_rate.csv
    - Входные признаки тестовой выборки: test_features.csv
    - Целевой признак тестовой выборки: test_target_job_satisfaction_rate.csv
- 2 задача:
    - Тренировочная выборка: train_quit.csv
    - Входные признаки тестовой выборки те же, что и в первой задаче: test_features.csv
    - Целевой признак тестовой выборки: test_target_quit.csv

Информация о признаках в данных:
- id — уникальный идентификатор сотрудника;
- dept — отдел, в котором работает сотрудник;
- level — уровень занимаемой должности;
- workload — уровень загруженности сотрудника;
- employment_years — длительность работы в компании (в годах);
- last_year_promo — показывает, было ли повышение за последний год;
- last_year_violations — показывает, нарушал ли сотрудник трудовой договор за последний год;
- supervisor_evaluation — оценка качества работы сотрудника, которую дал руководитель;
- salary — ежемесячная зарплата сотрудника;
- job_satisfaction_rate — уровень удовлетворённости сотрудника работой в компании, целевой признак.

## 1.2. Предобработка данных

Так как явные дубликаты в данных отсутствуют, названия столбцов и типы данных в них корректны, на этапе предобработки требуется **избавиться от пропусков** в данных и проверить таблицы на наличие **неявных дубликатов**.

### Работа с пропусками

Пропуски содержатся в столбцах dept и level датафреймов satisfaction_train и test_features. Пропусков совсем немного относительно общего объема данных, все они содержатся в категориальных признаках. В связи с этим, чтобы не исказить данные и не потерять информацию о пропусках, заменим пропуски в столбцах отдельной категорией 'unknown'.

Можно было бы попробовать заполнить попуски в столбце level на основе зарплаты сотрудника, так как обычно чем выше уровень сотрудника, тем выше у него зарплата. Однако, как можно заметить из первых строк датасетов, зарплата зависит не только от уровня, но также сильно определяется департаментом. Таким образом, пришлось бы учитывать одновременно несколько признаков для такого заполнения пропусков, что нецелесообразно при столь малом их числе.

```python
satisfaction_train['dept'] = satisfaction_train['dept'].replace(' ', np.nan)
satisfaction_train['level'] = satisfaction_train['level'].replace(' ', np.nan)


test_features['dept'] = test_features['dept'].replace(' ', np.nan)
test_features['level'] = test_features['level'].replace(' ', np.nan)
```

```python
print('Кол-во пропусков в датасете satisfaction_train:\n', satisfaction_train[['dept', 'level']].isna().sum())
print()
print('Кол-во пропусков в датасете test_features:\n', test_features[['dept', 'level']].isna().sum())
```

Неявные пропуски в датасетах были преобразованы в явные путем замены на  np.nan, чтобы в дальнейшем обработать пропуски в PipeLine.

### Неявные дубликаты

```python
def dubles_search(df, df_name):
    print(f'\033[1mУникальные значения в строковых столбцах датафрейма {df_name}:\033[0m')
    
    for column in df.select_dtypes(include=['object']).columns:
        print(f"\nСтолбец: '{column}'")
        print("Уникальные значения:")
        print(df[column].unique())
        print(f"Всего уникальных значений: {df[column].nunique()}")
        print()
```

```python
dubles_search(satisfaction_train, 'satisfaction_train')
dubles_search(test_features, 'test_features')
dubles_search(satisfaction_test_target, 'satisfaction_test_target')
```

```python
def compare_obj_plotting(df1, df2, column, figsize=(14, 7)):
    
    # Получаем имена датасетов из названий переменных
    df1_name = [k for k, v in globals().items() if v is df1][0]
    df2_name = [k for k, v in globals().items() if v is df2][0]
    
    colors = ['#FFD1DC', '#E0F7FA', '#D2F2E3', '#E1D5E7', '#F5D5CB', '#FFE6E6']
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
    
    # Первый график
    df1[column].value_counts().plot(
        kind='pie', 
        autopct='%1.1f%%',
        colors=colors,
        ax=ax1,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'},
        textprops={'fontsize': 10}
    )
    ax1.set_title(f'Распределение {column}\n{df1_name}', pad=20)
    ax1.set_ylabel('')
    
    # Второй график
    df2[column].value_counts().plot(
        kind='pie', 
        autopct='%1.1f%%',
        colors=colors,
        ax=ax2,
        wedgeprops={'linewidth': 1, 'edgecolor': 'white'},
        textprops={'fontsize': 10}
    )
    ax2.set_title(f'Распределение {column}\n{df2_name}', pad=20)
    ax2.set_ylabel('')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nСтатистика для {df1_name}:")
    print(df1[column].value_counts())
    print(f"\nСтатистика для {df2_name}:")
    print(df2[column].value_counts())
```

###  Признак 'employment_years'

```python
def discr_plotting_compare_target(df1, df2, column, target, figsize=(18, 12)):
    
    df1_name = [k for k, v in globals().items() if v is df1][0]
    df2_name = [k for k, v in globals().items() if v is df2][0]
    
    print(f"Распределения признака '{column}' в зависимости от '{target}' в двух датасетах:")
    
    fig, axes = plt.subplots(2, 2, figsize=figsize)
    
    # Верхние графики - обычные countplot
    sns.countplot(data=df1, x=column, hue=target, ax=axes[0, 0])
    axes[0, 0].set_title(f'Распределение ({column}) - {df1_name}')
    axes[0, 0].set_ylabel('Количество')
    axes[0, 0].tick_params(axis='x', rotation=45)
    axes[0, 0].legend(title=target)
    
    sns.countplot(data=df2, x=column, hue=target, ax=axes[0, 1])
    axes[0, 1].set_title(f'Распределение ({column}) - {df2_name}')
    axes[0, 1].set_ylabel('Количество')
    axes[0, 1].tick_params(axis='x', rotation=45)
    axes[0, 1].legend(title=target)
    
    # Нижние графики - нормированные barplot
    def get_percent_data(df, col, target):
        return (df.groupby([col, target])
                .size()
                .groupby(level=0)
                .apply(lambda x: 100 * x / x.sum())
                .reset_index(name='percent'))
    
    percent_df1 = get_percent_data(df1, column, target)
    percent_df2 = get_percent_data(df2, column, target)
    
    sns.barplot(data=percent_df1, x=column, y='percent', hue=target, ax=axes[1, 0])
    axes[1, 0].set_title(f'Нормированное распределение ({column}) - {df1_name}')
    axes[1, 0].set_ylabel('% сотрудников')
    axes[1, 0].tick_params(axis='x')
    axes[1, 0].legend(title=target)
    
    sns.barplot(data=percent_df2, x=column, y='percent', hue=target, ax=axes[1, 1])
    axes[1, 1].set_title(f'Нормированное распределение ({column}) - {df2_name}')
    axes[1, 1].set_ylabel('% сотрудников')
    axes[1, 1].tick_params(axis='x')
    axes[1, 1].legend(title=target)
    
    plt.tight_layout()
    plt.show()
    
    # Вывод статистики
    print(f"\nСтатистика для {df1_name}:")
    print(pd.crosstab(df1[column], df1[target], normalize='index'))
    print(f"\nСтатистика для {df2_name}:")
    print(pd.crosstab(df2[column], df2[target], normalize='index'))
```

###  Признак 'salary'

```python
def cont_plotting_compare(df1, df2, column, target=None, figsize=(18, 12)):
   
    df1_name = [k for k, v in globals().items() if v is df1][0]
    df2_name = [k for k, v in globals().items() if v is df2][0]
    
    fig, axes = plt.subplots(2, 2, figsize=figsize)
    
    # Верхние графики - нормированные гистограммы
    if target:
        # гистограммы с разбивкой по target
        sns.histplot(data=df1, x=column, hue=target, 
                    ax=axes[0, 0], kde=True, stat='density', common_norm=False)
        axes[0, 0].set_title(f'Нормированная гистограмма ({column})\n{df1_name}')
        axes[0, 0].set_ylabel('Плотность вероятности')
        
        sns.histplot(data=df2, x=column, hue=target, 
                    ax=axes[0, 1], kde=True, stat='density', common_norm=False)
        axes[0, 1].set_title(f'Нормированная гистограмма ({column})\n{df2_name}')
        axes[0, 1].set_ylabel('Плотность вероятности')
        
    else:
        # гистограммы без разбивки
        sns.histplot(df1, bins=20, ax=axes[0, 0], x=column, 
                    kde=True, stat='density', common_norm=False, color='skyblue')
        axes[0, 0].set_title(f'Нормированная гистограмма ({column})\n{df1_name}')
        
        sns.histplot(df2, bins=20, ax=axes[0, 1], x=column, 
                    kde=True, stat='density', common_norm=False, color='lightgreen')
        axes[0, 1].set_title(f'Нормированная гистограмма ({column})\n{df2_name}')
        
        row_offset = 0
    
    # Сравнение распределений
    sns.kdeplot(data=df1, x=column, ax=axes[1, 0], 
               color='skyblue', label=df1_name)
    sns.kdeplot(data=df2, x=column, ax=axes[1, 0], 
               color='lightgreen', label=df2_name)
    axes[1, 0].set_title(f'Сравнение распределений KDE\n({column})')
    axes[1, 0].set_ylabel('Плотность вероятности')
    axes[1, 0].legend()
    
    # Боксплоты
    combined_df = pd.concat([
        df1[[column]].assign(Dataset=df1_name),
        df2[[column]].assign(Dataset=df2_name)
    ])
    sns.boxplot(data=combined_df, y=column, x='Dataset', 
               ax=axes[1, 1], palette=['skyblue', 'lightgreen'])
    axes[1, 1].set_title(f'Сравнение квантилей\n({column})')
    axes[1, 1].set_ylabel(column)
    axes[1, 1].set_xlabel('')
    
    plt.tight_layout()
    plt.show()
    
    # Вывод статистики
    display(f"Описательная статистика для {df1_name}:")
    display(df1[column].describe())
    display(f"Описательная статистика для {df2_name}:")
    display(df2[column].describe())
    
    if target:
        display(f"Статистика по группам ({target}) для {df1_name}:")
        display(df1.groupby(target)[column].describe())
        display(f"Статистика по группам ({target}) для {df2_name}:")
        display(df2.groupby(target)[column].describe())
```

```python
cont_plotting_compare(satisfaction_train, df_test, 'salary', 'level')
```

Как можно было ожидать, меньше всего зарабатывают сотрудники уровня junior - обычно не больше 35-40тыс. Сотрудники уровня middle чаще всего зарабатывают от 40 до 60тыс. Зарплата сеньоров начинается от 70тыс. в обоих датасетах. Наблюдается резкое увеличение числа джуниоров с зарплатой 30 тыс - таких больше всего.

## 1.5. Обучение моделей

Возьмем для обучение 3 модели: LinearRegression, KNeighborsRegressor и DecisionTreeRegressor. Обучим их, подберем гиперпараметры и выберем лучшую. Для этого создадим общий пайплайн.

```python
pipe_final_1 = Pipeline(steps=[
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeRegressor(random_state=42))
])

param_grid_1 = [
    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsRegressor() 
    {
        'models': [KNeighborsRegressor()],
        'models__n_neighbors': range(2,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели DecisionTreeRegressor()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_features': range(2,10),
        'models__min_samples_split': range(2,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]
```

Для оценки качества лучшей модели будем использовать метрику SMAPE («симметричное среднее абсолютное процентное отклонение»). Она вычисляется по следующей формуле:

$$
\text{SMAPE} = \frac{100}{n} \sum_{i=1}^n \frac{|y_i - \hat{y}_i|}{(|y_i| + |\hat{y}_i|)/2}
$$

**где:**

- $y_i$ — фактическое значение целевого признака для объекта с порядковым номером $i$ в выборке;
- $\hat{y}_i$ — предсказанное значение целевого признака для объекта с порядковым номером $i$ в выборке;
- $n$ — количество объектов в выборке;
- $\sum_{i=1}^n$ — сумма значений, полученная в результате операций, которые следуют за этим знаком, для всех объектов с порядковым номером от $i$ до $n$ в выборке.

Для использования SMAPE создадим собственную метрику:

```python
# Создаем собственную метрику SMAPE
def smape(y_true, y_pred):
    return 100 * np.mean(2 * np.abs(y_true - y_pred) / (np.abs(y_true) + np.abs(y_pred)))
```

```python
smape_scorer = make_scorer(smape, greater_is_better=False) 
```

```python
grid_search_1 = GridSearchCV(
    pipe_final_1,
    param_grid_1,
    cv=5,
    scoring=smape_scorer,
    n_jobs=-1,
    error_score='raise'
)
grid_search_1.fit(X_train_1, y_train_1)

print('Лучшая модель и её параметры:\n\n', grid_search_1.best_estimator_)
print ('Метрика SMAPE лучшей модели на кросс-валидации:', np.abs(grid_search_1.best_score_))
```

```python
X_test_1 = df_test.drop(['job_satisfaction_rate', 'id'], axis=1)
y_test_1 = df_test['job_satisfaction_rate']
y_pred = grid_search_1.predict(X_test_1)
```

```python
test_smape = smape(y_test_1, y_pred)
print(f'SMAPE на тесте: {test_smape:.2f}%')
```

Это также отлично видно на нормированной гистограмме.

Чтобы выполнить оценку более статистически достоверно, воспользуемся статистическим тестом - независимым (двухвыборочным) t-тестом о равенстве средних двух генеральных совокупностей:

```python
array_no = df_test_full[df_test_full['quit'] == 'no']['job_satisfaction_rate']
array_yes = df_test_full[df_test_full['quit'] == 'yes']['job_satisfaction_rate']
```

- Нулевая гипотеза **(H0)**: Средняя удовлетворенность уволившихся и оставшихся сотрудников равна в пределах допустимого.
- Альтернативная гипотеза **(H1)**: Средняя удовлетворенность сильно отличается для 2 групп. Таким образом, увольнение сотрудника зависит от его удовлетворенности.

```python
alpha = 0.05 # уровень статистической значимости

results = st.ttest_ind(array_no, array_yes, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Таким образом, мы статистически доказали, что уровень удовлетворенности сотрудника на работе влияет на его возможное увольнение.

## 2.6. Обучение моделей

```python
ohe_pipe = Pipeline(
    [
        (
            'imputer', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
        )
    ]
)


ord_pipe = Pipeline(
    [
        (
            'imputer', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ord',  
            OrdinalEncoder(
                categories=[
                    ['junior', 'middle', 'senior'], 
                    ['low', 'medium', 'high']
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
        (
            'imputer_after', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        )
    ]
) 

# общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, cols_ohe),
        ('ord', ord_pipe, cols_ord),
        ('num', MinMaxScaler(), cols_num)
    ], 
    remainder='passthrough'
)

pipe_final_2 = Pipeline(steps=[
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])

param_grid_2 = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_features': range(2,5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,7),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(penalty='l1', random_state=RANDOM_STATE)],
        'models__C': [0.2, 1, 10],
        'models__solver': ['liblinear', 'saga'],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели SVC()
    {
        'models': [SVC(probability=True, random_state=RANDOM_STATE)],
        'models__kernel': ['rbf', 'linear'],
        'models__C': [2, 7, 10],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]
```

```python
randomized_search_2 = RandomizedSearchCV(
    pipe_final_2, 
    param_grid_2, 
    cv=5,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)
randomized_search_2.fit(X_train_2, y_train_2)

print('Лучшая модель и её параметры:\n\n', randomized_search_2.best_estimator_)
print ('Метрика ROC-AUC лучшей модели на кросс-валидации:', randomized_search_2.best_score_)
```

```python
y_pred_2 = randomized_search_2.predict(X_test_2)
y_pred_2_proba = randomized_search_2.predict_proba(X_test_2)[:, 1] 
```

```python
test_roc_auc = roc_auc_score(y_test_2, y_pred_2_proba)
print(f'ROC-AUC на тесте: {test_roc_auc:.2f}')
```
